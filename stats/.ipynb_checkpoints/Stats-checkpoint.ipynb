```python
import pandas as pd
```

```python
file_path = "../dataset/keypoint_angle_outputs/all_pose_angles.csv"
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv(file_path)

# Define a custom function to calculate the 25th and 75th percentiles
def q1(x):
    return x.quantile(0.25)

def q3(x):
    return x.quantile(0.75)

# Group by 'pose' and calculate statistics
grouped_stats = df.groupby('pose').agg(['mean', 'std', 'min', q1, 'median', q3, 'max'])
```
